```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import neighbors
from sklearn.model_selection import train_test_split 
%matplotlib notebook
```

```python
cancer =  datasets.load_breast_cancer()
cancer_df =  pd.DataFrame(cancer.data, columns =  cancer.feature_names)
cancer_df['target'] =  cancer.target
```

```python
X =  cancer_df[cancer.feature_names]
y =  cancer_df['target']
```

```python
X_train,X_test,y_train,y_test =  train_test_split(X,y,test_size = 0.25,random_state = 0)

score = []
nearest_n = []

for k in range(1,100):
    KNN = neighbors.KNeighborsClassifier(n_neighbors = k)
    KNN.fit(X_train,y_train)
    prediction_score =  KNN.score(X_test,y_test)
    score.append(prediction_score)
    nearest_n.append(k)
    
    
```

```python
plt.plot(nearest_n,score)
plt.scatter(nearest_n,score)

plt.xlabel("K")
plt.ylabel("Score")
plt.xticks(range(0,105,10))
plt.title("K values from 1-100 for a KNN Model")

plt.show()
```
